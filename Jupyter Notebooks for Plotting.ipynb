<div style="background-color: #cce5ff; color: #004085; padding: 20px; border-radius: 5px; margin-top: 20px; border: 1px solid #b8daff;">Displaying Charts in Separate Cells</div>

```python
# Displaying Multiple Plots in Jupyter Notebooks

import matplotlib.pyplot as plt
import numpy as np

# First plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Second plot
y2 = np.cos(x)

plt.figure()
plt.plot(x, y2)
plt.title('Cosine Wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

<div style="background-color: #cce5ff; color: #004085; padding: 20px; border-radius: 5px; margin-top: 20px; border: 1px solid #b8daff;">Displaying Multiple Charts in a Single Cell</div>

```python
#Displaying Multiple Plots in Jupyter Charts in a Single Cell

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# First plot
y1 = np.sin(x)
plt.figure()
plt.plot(x, y1)
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Second plot
y2 = np.cos(x)
plt.figure()
plt.plot(x, y2)
plt.title('Cosine Wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

<div style="background-color: #cce5ff; color: #004085; padding: 20px; border-radius: 5px; margin-top: 20px; border: 1px solid #b8daff;">Using Subplots to Display Multiple Charts in one Figure</div>

```python
#Using Subplots to Display Multiple Charts in One Figure

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# First subplot
axs[0].plot(x, y1)
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Second subplot
axs[1].plot(x, y2)
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```
