```python
import matplotlib.pyplot as plt

# Stages and data
stages = [
    "Intelligent\nAutomation",
    "LLM Copilots\nin Workflows",
    "Agentic AI\nChatbots",
    "Multi-Agent\nEcosystems",
    "Proto-AGI\nWorkflows",
    "AGI-Enabled\nChatbots"
]

# User AI knowledge to use (scale 1-10)
knowledge_use = [2, 2, 3, 4, 4, 3]

# Relative AI power (orders of magnitude 10 to 10)
power = [10**1, 10**2, 10**3, 10**4, 10**5, 10**6]

fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot user knowledge (primary y-axis)
ax1.plot(stages, knowledge_use, marker='o', label="AI Knowledge Needed (110)", color='tab:blue')
ax1.set_ylim(0, 11)
ax1.set_ylabel("AI Knowledge Needed (110)", color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Annotate points for knowledge
for x, y in zip(stages, knowledge_use):
    ax1.text(x, y + 0.3, f"{y}", ha='center', color='tab:blue')

# Secondary axis for AI power (log scale)
ax2 = ax1.twinx()
ax2.plot(stages, power, marker='s', label="AI Power (log scale)", color='tab:orange')
ax2.set_yscale('log')
ax2.set_ylabel("Relative AI Power (log)", color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Annotate points for power
for x, y in zip(stages, power):
    ax2.text(x, y*1.2, f"{int(y):,}", ha='center', color='tab:orange')

# Title and grid
plt.title("AI Knowledge of User and How Powerful is the AI")
ax1.grid(True, axis='y', linestyle='--', alpha=0.4)

# Legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# vertical line at 3.75, make is shaded and label it as "AI Knowledge Threshold"

ax1.axvline(x=2.75, color='gray', linestyle='--', label='AI Knowledge Threshold')
ax1.fill_betweenx([0, 11], 2.75, 3.25, alpha=0.1, color='gray')



plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Load the Iris dataset
data = load_iris()
X = data.data         # Features (4D)
y = data.target       # Labels (0 = setosa, 1 = versicolor, 2 = virginica)
target_names = data.target_names

# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 4: Plot the PCA results
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], alpha=0.7, label=target_name, color=colors[i])

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5: Print explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
# Factor Analysis Example with Promax Rotation
#!/usr/bin/env python3
# This script demonstrates how to perform Factor Analysis on the Iris dataset using Promax rotation.


# First, we need to install the 'factor_analyzer' library
# This library is not included in scikit-learn and needs to be installed separately.
print("Installing factor_analyzer...")
%pip install factor_analyzer
print("Installation complete.")

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer

# --- 1. Load and Prepare the Data ---
# Load the Iris dataset
print("\nLoading Iris dataset...")
iris = load_iris()
# Create a pandas DataFrame for easier handling
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print("Dataset loaded successfully. Features:", list(df.columns))

# --- 2. Perform Factor Analysis ---
# We will run the analysis with 2 factors and a 'promax' (oblique) rotation.
print("\nPerforming Factor Analysis with 2 factors and promax rotation...")
fa = FactorAnalyzer(n_factors=2, rotation='promax', method='principal')
fa.fit(df)
print("Factor analysis complete.")

# --- 3. Get and Display Factor Loadings ---
# Factor loadings represent how much a variable is correlated with a factor.
loadings = fa.loadings_
loadings_df = pd.DataFrame(loadings, index=df.columns, columns=['Factor 1', 'Factor 2'])

print("\n--- Factor Loadings ---")
# Using .style.background_gradient() makes the table easier to read by highlighting larger values.
display(loadings_df.style.background_gradient(cmap='viridis', axis=None))


# --- 4. Get and Display Explained Variance ---
# This tells us how much of the total variance in the original data is captured by each factor.
variance = fa.get_factor_variance()
variance_df = pd.DataFrame(
    variance,
    index=['Sum of Squared Loadings (Eigenvalue)', 'Proportional Variance', 'Cumulative Variance'],
    columns=['Factor 1', 'Factor 2']
)

print("\n--- Explained Variance ---")
display(variance_df)

# --- 5. Visualize the Factor Loadings ---
print("\nGenerating plot of factor loadings...")
# Create a bar plot to visualize the loadings of each feature on the two factors.
ax = loadings_df.plot(kind='bar', figsize=(12, 7), rot=0)

# Set plot titles and labels for clarity
ax.set_title('Factor Loadings with Promax Rotation', fontsize=16)
ax.set_xlabel('Features', fontsize=12)
ax.set_ylabel('Factor Loading', fontsize=12)
ax.axhline(0, color='black', linewidth=0.5) # Add a zero line for reference
plt.legend(title='Factors')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
print("Plot generated.")

#create a 3D scatter plot of the factor scores
# 3D Scatter Plot of Factor Scores

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Generate factor scores
factor_scores = fa.transform(df)
ax.scatter(factor_scores[:, 0], factor_scores[:, 1], zs=0, zdir='z', s=50, alpha=0.7, c='blue')
ax.set_xlabel('Factor 1')
ax.set_ylabel('Factor 2')
ax.set_zlabel('Factor 3 (not used)')
ax.set_title('3D Scatter Plot of Factor Scores')
plt.show()
# Display the factor scores in a DataFrame
factor_scores_df = pd.DataFrame(factor_scores, columns=['Factor 1', 'Factor 2'])
print("\n--- Factor Scores ---")
display(factor_scores_df.head())
# Display the factor scores in a DataFrame
print("\n--- Factor Scores ---")
display(factor_scores_df.head())
# Display the factor scores in a DataFrame
print("\n--- Factor Scores ---")
```
