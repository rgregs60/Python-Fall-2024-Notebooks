```python
### Let fix Iris Dataset
%reset -f 
# ! pip install --upgrade kaleido
import plotly.express as px
import plotly.graph_objects as go
import pandas
import numpy as np

# Load the Iris dataset
Iris_Data = px.data.iris()

# Group by species and calculate the mean and standard error of sepal_width
Iris_group = Iris_Data.groupby('species').agg(
    mean_sepal_width=('sepal_width', 'mean'),
    se_sepal_width=('sepal_width', lambda x: np.std(x) / np.sqrt(len(x)))
).reset_index()

# Calculate the 95% confidence interval
Iris_group['ci_sepal_width'] = 1.96 * Iris_group['se_sepal_width']

Iris_group['mean_sepal_width'] = Iris_group['mean_sepal_width'].apply(lambda x: round(x, 2))

# Create a bar chart with error bars
fig = px.bar(
    Iris_group,
    x='species',
    y='mean_sepal_width',
    error_y='ci_sepal_width',
    title='Average Sepal Width by Species with 95% Confidence Interval',
    labels={'mean_sepal_width': 'Average Sepal Width'}
)

# Format the y-axis to have two decimal places
fig.update_layout(
    yaxis=dict(
        tickformat=".0f"
    )
)
fig.show()

# Group by species and calculate the mean and standard error of sepal_length
Iris_group = Iris_Data.groupby('species').agg(
    mean_sepal_length=('sepal_length', 'mean'),
    se_sepal_length=('sepal_length', lambda x: np.std(x) / np.sqrt(len(x)))
).reset_index()

# Calculate the 95% confidence interval
Iris_group['ci_sepal_length'] = 1.96 * Iris_group['se_sepal_length']

Iris_group['mean_sepal_length'] = Iris_group['mean_sepal_length'].apply(lambda x: round(x, 2))

# Create a bar chart with error bars
fig = px.bar(
    Iris_group,
    x='species',
    y='mean_sepal_length',
    error_y='ci_sepal_length',
    title='Average Sepal Length by Species with 95% Confidence Interval',
    labels={'mean_sepal_length': 'Average Sepal Length'}
)

# Format the y-axis to have two decimal places
fig.update_layout(
    yaxis=dict(
        tickformat=".0f"
    )
)

fig.show()
```
