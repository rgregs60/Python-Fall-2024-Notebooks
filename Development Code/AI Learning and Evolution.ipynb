```python
import matplotlib.pyplot as plt

# Stages and data
stages = [
    "Intelligent\nAutomation",
    "LLM Copilots\nin Workflows",
    "Agentic AI\nChatbots",
    "Multi-Agent\nEcosystems",
    "Proto-AGI\nWorkflows",
    "AGI-Enabled\nChatbots"
]

# User AI knowledge to use (scale 1-10)
knowledge_use = [2, 2, 3, 4, 4, 3]

# Relative AI power (orders of magnitude 10 to 10)
power = [10**1, 10**2, 10**3, 10**4, 10**5, 10**6]

fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot user knowledge (primary y-axis)
ax1.plot(stages, knowledge_use, marker='o', label="AI Knowledge Needed (110)", color='tab:blue')
ax1.set_ylim(0, 11)
ax1.set_ylabel("AI Knowledge Needed (110)", color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Annotate points for knowledge
for x, y in zip(stages, knowledge_use):
    ax1.text(x, y + 0.3, f"{y}", ha='center', color='tab:blue')

# Secondary axis for AI power (log scale)
ax2 = ax1.twinx()
ax2.plot(stages, power, marker='s', label="AI Power (log scale)", color='tab:orange')
ax2.set_yscale('log')
ax2.set_ylabel("Relative AI Power (log)", color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Annotate points for power
for x, y in zip(stages, power):
    ax2.text(x, y*1.2, f"{int(y):,}", ha='center', color='tab:orange')

# Title and grid
plt.title("AI Knowledge of User and How Powerful is the AI")
ax1.grid(True, axis='y', linestyle='--', alpha=0.4)

# Legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# vertical line at 3.75, make is shaded and label it as "AI Knowledge Threshold"

ax1.axvline(x=2.75, color='gray', linestyle='--', label='AI Knowledge Threshold')
ax1.fill_betweenx([0, 11], 2.75, 3.25, alpha=0.1, color='gray')



plt.tight_layout()
plt.show()
```
