```python
import pandas as pd

data = [
    {"PSCDI": 9,  "c29": 34.1968, "logk": -5.5229},
    {"PSCDI": 12, "c29": 34.1968, "logk": -4.0721},
    {"PSCDI": 15, "c29": 34.1968, "logk": -2.6214},
    {"PSCDI": 9,  "c29": 59.79,   "logk": -4.6513},
    {"PSCDI": 12, "c29": 59.79,   "logk": -4.0062},
    {"PSCDI": 15, "c29": 59.79,   "logk": -3.361},
    {"PSCDI": 9,  "c29": 91.8336, "logk": -3.5601},
    {"PSCDI": 12, "c29": 91.8336, "logk": -3.9236},
    {"PSCDI": 15, "c29": 91.8336, "logk": -4.287},
]

dfGaby = pd.DataFrame(data)

# Loaded variable 'dfGaby' from kernel state

# Rename index column to 'Record ID'
dfGaby.index.name = 'Record ID'

print(dfGaby)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=dfGaby, x="c29", y="logk", hue="PSCDI", marker="o")

plt.title("logk vs c29 grouped by PSCDI")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style("white")

# Create the lineplot
ax = sns.lineplot(
    data=dfGaby,
    x="c29",
    y="logk",
    hue="PSCDI",
    marker="o"
)

# --- Axis formatting ---
ax.set_title("logk vs c29 grouped by PSCDI", fontsize=14, weight="bold")
ax.set_xlabel("c29", fontsize=12)

# Left Y-axis labeled as K
ax.set_ylabel("K", fontsize=12)

# Move y-axis ticks and labels to the right
ax.yaxis.tick_right()
ax.yaxis.set_label_position("left")   # label stays on the left
ax.invert_yaxis()                     # reverse order

# --- Add value labels at the end of each line ---
for ps in dfGaby["PSCDI"].unique():
    subset = dfGaby[dfGaby["PSCDI"] == ps]
    x_val = subset["c29"].iloc[-1]
    y_val = subset["logk"].iloc[-1]
    ax.text(
        x_val + 1, y_val, f"{y_val:.2f}",  # place slightly to the right
        va="center", ha="left"
    )

plt.legend(title="PSCDI", loc="best")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Use a clean style
sns.set_style("white")

# Create the line plot with c29 on the Y-axis
ax = sns.lineplot(
    data=dfGaby,
    x="logk",      # logk on X
    y="c29",       # c29 on Y
    hue="PSCDI",
    marker=None,   # remove markers
    linewidth=2
)

# --- Axis formatting ---
ax.set_title("logk vs c29 grouped by PSCDI", fontsize=14, weight="bold")
ax.set_xlabel("logk", fontsize=12)
ax.set_ylabel("c29", fontsize=12)

# Y-axis on the right side, reversed
ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")
ax.invert_yaxis()

# Remove gridlines
sns.despine(left=True, bottom=True)

# --- Remove borders (spines) completely ---
for spine in ax.spines.values():
    spine.set_visible(False)

plt.legend(title="PSCDI", loc="best")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure
fig, ax = plt.subplots(figsize=(6,6))

# Draw Earth (circle)
earth = patches.Circle((0,0), radius=1, color="lightblue", ec="blue", alpha=0.5)
ax.add_patch(earth)

# Add East (Sunrise) and West (Sunset) labels with sun graphics
ax.text(1.3, 0, " Sunrise\n(East)", ha="left", va="center", fontsize=12, color="orange")
ax.text(-1.3, 0, " Sunset\n(West)", ha="right", va="center", fontsize=12, color="red")

# Place "person" stick figure on the west side
ax.plot(-1, 0, marker="o", markersize=15, color="black")  # head
ax.plot([-1, -1], [-0.1, -0.4], color="black", linewidth=2)  # body
ax.plot([-1, -1.2], [-0.15, -0.25], color="black", linewidth=2)  # left arm
ax.plot([-1, -0.8], [-0.15, -0.25], color="black", linewidth=2)  # right arm
ax.plot([-1, -1.1], [-0.4, -0.6], color="black", linewidth=2)  # left leg
ax.plot([-1, -0.9], [-0.4, -0.6], color="black", linewidth=2)  # right leg

# Add explanatory text
ax.text(0, -1.3, "Closest to Sunset (West)\nFarthest from Sunrise (East)", 
        ha="center", va="top", fontsize=11, color="blue")

# Formatting
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_aspect("equal")
ax.axis("off")

plt.show()
```
