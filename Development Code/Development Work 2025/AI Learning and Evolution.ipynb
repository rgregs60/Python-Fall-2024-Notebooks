```python
# Conceptual example of an agent using an LLM to generate a tool (function)

class SelfImprovingAgent:
    def __init__(self, llm_code_generator):
        self.llm_code_generator = llm_code_generator
        self.tools = {} # A dictionary to store dynamically created tools

    def _generate_code_with_llm(self, prompt_for_tool_code):
        """
        Conceptual function to call a Large Language Model for code generation.
        In a real system, this would involve API calls to a model like GPT-4,
        Claude, Gemini, etc., specifically prompted for code.
        """
        print(f"Agent: Sending prompt to LLM: '{prompt_for_tool_code}'")
        # This is a placeholder for actual LLM API call and response handling
        # For example, if the prompt was "Write a Python function named 'add' that takes two numbers and returns their sum."
        # The LLM might return: "def add(a, b):\n  return a + b"
        generated_code_str = self.llm_code_generator.generate(prompt_for_tool_code) # Conceptual call
        print(f"Agent: Received code from LLM:\n{generated_code_str}")
        return generated_code_str

    def _integrate_and_test_tool(self, tool_name, tool_code_str):
        """
        Conceptual function to dynamically integrate and test the new tool.
        This is a highly simplified example. Real integration is complex and risky.
        """
        try:
            # Attempt to execute the generated code to define the function
            # In a safer environment, this would use sandboxing and careful parsing.
            namespace = {}
            exec(tool_code_str, namespace)
            
            if tool_name not in namespace:
                print(f"Agent: Error - Tool '{tool_name}' not found in generated code's namespace.")
                return False

            new_tool_function = namespace[tool_name]
            
            # Basic conceptual test (highly simplified)
            # A real system would have a robust testing framework based on the tool's purpose.
            print(f"Agent: Testing tool '{tool_name}'...")
            if tool_name == "greet_user": # Example specific test
                if callable(new_tool_function) and new_tool_function("Alice") == "Hello, Alice!":
                    self.tools[tool_name] = new_tool_function
                    print(f"Agent: Tool '{tool_name}' integrated and passed basic test.")
                    return True
                else:
                    print(f"Agent: Tool '{tool_name}' failed basic test.")
                    return False
            else:
                # For other tools, a more generic test or no test in this simple example
                self.tools[tool_name] = new_tool_function
                print(f"Agent: Tool '{tool_name}' integrated (assuming basic functionality).")
                return True
        except Exception as e:
            print(f"Agent: Error integrating or testing tool '{tool_name}': {e}")
            return False

    def create_new_tool(self, tool_name, tool_description_for_llm):
        """
        High-level function for the agent to attempt to create a new tool.
        """
        print(f"\nAgent: Attempting to create a new tool: '{tool_name}'")
        prompt = f"Write a Python function named '{tool_name}' that {tool_description_for_llm}."
        
        generated_code = self._generate_code_with_llm(prompt)
        
        if generated_code:
            self._integrate_and_test_tool(tool_name, generated_code)
        else:
            print("Agent: Failed to generate code for the tool.")

    def use_tool(self, tool_name, *args):
        if tool_name in self.tools:
            return self.tools[tool_name](*args)
        else:
            return f"Tool '{tool_name}' not available."

# --- Conceptual LLM Code Generator (Placeholder) ---
class MockLLMCodeGenerator:
    def generate(self, prompt):
        # Simulate LLM response based on prompt
        if "greet_user" in prompt and "takes a name and returns a greeting" in prompt:
            return "def greet_user(name):\n  return f'Hello, {name}!'"
        elif "add_numbers" in prompt and "takes two numbers and returns their sum" in prompt:
            return "def add_numbers(x, y):\n  return x + y"
        return "# LLM failed to generate code for this prompt"

# --- Example Usage ---
if __name__ == "__main__":
    # Initialize the agent with a mock LLM code generator
    agent = SelfImprovingAgent(llm_code_generator=MockLLMCodeGenerator())

    # Agent decides it needs a tool to greet users
    agent.create_new_tool(
        tool_name="greet_user",
        tool_description_for_llm="takes a name as a string and returns a greeting string like 'Hello, [name]!'"
    )

    # Agent decides it needs a tool to add numbers
    agent.create_new_tool(
        tool_name="add_numbers",
        tool_description_for_llm="takes two numbers and returns their sum"
    )

    # Agent uses its newly created tool
    greeting = agent.use_tool("greet_user", "Bob")
    print(f"Agent using tool: {greeting}") # Expected: Hello, Bob!

    sum_result = agent.use_tool("add_numbers", 5, 7)
    print(f"Agent using tool: {sum_result}") # Expected: 12
    
    unknown_tool_result = agent.use_tool("subtract_numbers", 10, 3)
    print(f"Agent using tool: {unknown_tool_result}")
```

```python
# Factor Analysis Example with Oblimin Rotation
#!/usr/bin/env python3
# This script demonstrates how to perform Factor Analysis on the Iris dataset using Oblimin rotation.


# First, we need to install the 'factor_analyzer' library
# This library is not included in scikit-learn and needs to be installed separately.

print("Installing factor_analyzer...")
%pip install factor_analyzer
print("Installation complete.")

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer

# --- 1. Load and Prepare the Data ---
# Load the Iris dataset
print("\nLoading Iris dataset...")
iris = load_iris()
# Create a pandas DataFrame for easier handling
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print("Dataset loaded successfully. Features:", list(df.columns))

# --- 2. Perform Factor Analysis ---
# We will run the analysis with 2 factors and a 'oblimin' (oblique) rotation.
print("\nPerforming Factor Analysis with 2 factors and oblimin rotation...")
fa = FactorAnalyzer(n_factors=2, rotation='oblimin', method='principal')
fa.fit(df)
print("Factor analysis complete.")

# --- 3. Get and Display Factor Loadings ---
# Factor loadings represent how much a variable is correlated with a factor.
loadings = fa.loadings_
loadings_df = pd.DataFrame(loadings, index=df.columns, columns=['Factor 1', 'Factor 2'])

print("\n--- Factor Loadings ---")
# Using .style.background_gradient() makes the table easier to read by highlighting larger values.
display(loadings_df.style.background_gradient(cmap='viridis', axis=None))


# --- 4. Get and Display Explained Variance ---
# This tells us how much of the total variance in the original data is captured by each factor.
variance = fa.get_factor_variance()
variance_df = pd.DataFrame(
    variance,
    index=['Sum of Squared Loadings (Eigenvalue)', 'Proportional Variance', 'Cumulative Variance'],
    columns=['Factor 1', 'Factor 2']
)

print("\n--- Explained Variance ---")
display(variance_df)

# --- 5. Visualize the Factor Loadings ---
print("\nGenerating plot of factor loadings...")
# Create a bar plot to visualize the loadings of each feature on the two factors.
ax = loadings_df.plot(kind='bar', figsize=(12, 7), rot=0)

# Set plot titles and labels for clarity
ax.set_title('Factor Loadings with Oblimin Rotation', fontsize=16)
ax.set_xlabel('Features', fontsize=12)
ax.set_ylabel('Factor Loading', fontsize=12)
ax.axhline(0, color='black', linewidth=0.5) # Add a zero line for reference
plt.legend(title='Factors')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
print("Plot generated.")

#create a 3D scatter plot of the factor scores
# 3D Scatter Plot of Factor Scores

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Generate factor scores
factor_scores = fa.transform(df)
ax.scatter(factor_scores[:, 0], factor_scores[:, 1], zs=0, zdir='z', s=50, alpha=0.7, c='blue')
ax.set_xlabel('Factor 1')
ax.set_ylabel('Factor 2')
ax.set_zlabel('Factor 3 (not used)')
ax.set_title('3D Scatter Plot of Factor Scores')
plt.show()
# Display the factor scores in a DataFrame
factor_scores_df = pd.DataFrame(factor_scores, columns=['Factor 1', 'Factor 2'])
print("\n--- Factor Scores ---")
display(factor_scores_df.head())
# Display the factor scores in a DataFrame
print("\n--- Factor Scores ---")
display(factor_scores_df.head())
# Display the factor scores in a DataFrame
print("\n--- Factor Scores ---")

# make the 3d factor score plot interactive using plotly

import plotly.express as px

#create an interactive 3D scatter plot of the factor scores color code the observations by their species
# Create a DataFrame for factor scores with species labels

factor_scores_df = pd.DataFrame(factor_scores, columns=['Factor 1', 'Factor 2'])
factor_scores_df['Species'] = pd.Categorical.from_codes(iris.target, iris.target_names) 

# Create an interactive 3D scatter plot of the factor scores using Plotly
fig = px.scatter_3d(factor_scores_df, x='Factor 1', y='Factor 2', z=factor_scores_df.index,
                    title='Interactive 3D Scatter Plot of Factor Scores',
                    labels={'index': 'Sample Index'})
fig.update_traces(marker=dict(size=5, color=factor_scores_df['Species'].cat.codes, opacity=0.7))
fig.update_layout(scene=dict(
    xaxis_title='Factor 1',
    yaxis_title='Factor 2',
    zaxis_title='Sample Index'
))

# Create a color bar for the species
fig.update_traces(marker=dict(color=factor_scores_df['Species'].cat.codes, colorscale='Viridis', showscale=True))


fig.show()

# Display the factor scores in a DataFrame
print("\n--- Factor Scores ---")
display(factor_scores_df.head())
# Display the factor scores in a DataFrame
```

```python
#### Technology Adoption S-Curve

import numpy as np
import matplotlib.pyplot as plt

# Time axis (arbitrary units)
time = np.linspace(0, 20, 200)

# Logistic (S-curve) function for diffusion
def logistic(t, L=1, k=0.5, x0=10):
    return L / (1 + np.exp(-k*(t - x0)))

# Adoption curve (cumulative % of market)
adoption_curve = logistic(time, L=100, k=0.6, x0=10)

# Rate of adoption (derivative of logistic) - bell curve
rate_curve = (0.6*100*np.exp(-0.6*(time-10))) / ((1+np.exp(-0.6*(time-10)))**2)

# Plot
plt.figure(figsize=(10,6))

# Plot cumulative adoption (S-curve)
plt.plot(time, adoption_curve, label="Cumulative Adoption (S-curve)", color="blue", linewidth=2)

# Plot rate of adoption (bell curve)
plt.plot(time, rate_curve, label="Rate of Adoption (Bell Curve)", color="orange", linestyle="--", linewidth=2)

# Labels
plt.title("Technology Diffusion Over Time", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Percent of Market Adoption", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
### S-Curve of Technology Adoption, Inflection Point, and Adopter Categories

import numpy as np
import matplotlib.pyplot as plt

# Time axis
time = np.linspace(0, 20, 400)

# Logistic function for cumulative adoption
def logistic(t, L=1, k=0.6, x0=10):
    return L / (1 + np.exp(-k*(t - x0)))

# Adoption curve (cumulative % of market)
adoption_curve = logistic(time, L=100, k=0.6, x0=10)

# Inflection point (where diminishing returns begin)
inflection_x = 10
inflection_y = logistic(inflection_x, L=100, k=0.6, x0=10)

# Adopter categories (% of population)
categories = {
    "Innovators": 2.5,
    "Early Adopters": 13.5,
    "Early Majority": 34,
    "Late Majority": 34,
    "Laggards": 16
}

# Cumulative category boundaries
boundaries = np.cumsum(list(categories.values()))

# Map category midpoints to time (approximate using inverse logistic)
def inv_logistic(y, L=100, k=0.6, x0=10):
    return x0 - (1/k) * np.log(L/y - 1)

midpoints = [inv_logistic(b - (p/2), L=100, k=0.6, x0=10) 
             for b, p in zip(boundaries, categories.values())]

# Plot
plt.figure(figsize=(12,7))

# Plot adoption S-curve
plt.plot(time, adoption_curve, color="blue", linewidth=2, label="Cumulative Adoption (S-curve)")

# Inflection point (dotted vertical line)
plt.axvline(x=inflection_x, color="black", linestyle=":", linewidth=1.5, label="Inflection Point")

# Annotate categories
for (cat, pct), mid in zip(categories.items(), midpoints):
    y_val = logistic(mid, L=100, k=0.6, x0=10)
    plt.text(mid, y_val+3, cat, ha="center", fontsize=10, fontweight="bold")

# Labels and title
plt.title("Technology Diffusion: S-Curve with Adopter Categories", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Cumulative Market Adoption (%)", fontsize=12)
plt.ylim(0, 105)
plt.xlim(0, 20)
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Consumer vs Enterprise-Scale Generative AI Placement on the Diffusion S-Curve (2025)

import numpy as np
import matplotlib.pyplot as plt

# --- Logistic S-curve setup ---
time = np.linspace(0, 20, 500)

def logistic(t, L=100, k=0.6, x0=10):
    return L / (1 + np.exp(-k*(t - x0)))

def inv_logistic(y, L=100, k=0.6, x0=10):
    return x0 - (1/k) * np.log(L/y - 1)

curve = logistic(time)

# Category cumulative boundaries (%): Innovators 2.5, Early Adopters 16, Early Majority 50, Late Majority 84
boundaries_pct = [2.5, 16, 50, 84, 100]
boundary_x = [inv_logistic(p) for p in boundaries_pct]

# Category midpoint x-positions for labels
midpoints_pct = [(0+2.5)/2, (2.5+16)/2, (16+50)/2, (50+84)/2, (84+100)/2]
mid_x = [inv_logistic(p) for p in midpoints_pct]
cat_names = ["Innovators", "Early Adopters", "Early Majority", "Late Majority", "Laggards"]

# Placement per corrected request:
# Consumer GenAI: firmly in Early Majority, but not yet at inflection (3035% adoption)
consumer_x = inv_logistic(33)  
consumer_y = logistic(consumer_x)

# Enterprise-scale GenAI: Early Adopters (~12% adoption)
enterprise_x = inv_logistic(12)  
enterprise_y = logistic(enterprise_x)

# --- Plot ---
plt.figure(figsize=(12,7))

# S-curve
plt.plot(time, curve, linewidth=2, label="Cumulative Adoption (S-curve)")

# Inflection point (50%)
plt.axvline(x=boundary_x[2], linestyle=":", linewidth=1.5, label="Inflection Point (50%)")

# Category boundary markers (light dashed)
for x in boundary_x[:-1]:  # skip 100%
    plt.axvline(x=x, linestyle="--", linewidth=0.8, alpha=0.4)

# Category labels
for name, x in zip(cat_names, mid_x):
    y = logistic(x)
    plt.text(x, y+3, name, ha="center", fontsize=10, fontweight="bold")

# User-specified markers
plt.axvline(x=consumer_x, color="purple", linewidth=4, label="Consumer GenAI (2025)  Early Majority")
plt.axvline(x=enterprise_x, color="darkgreen", linewidth=4, label="Enterprise GenAI (2025)  Early Adopters")

# Annotations
plt.text(consumer_x+0.15, consumer_y-6, "Consumer GenAI  2025", rotation=90, color="purple", fontsize=11, fontweight="bold")
plt.text(enterprise_x+0.15, enterprise_y-6, "Enterprise GenAI  2025", rotation=90, color="darkgreen", fontsize=11, fontweight="bold")

# Titles and axes
plt.title("Technology Diffusion S-curve (Logistic) with Adopter Categories\nConsumer vs Enterprise-Scale Generative AI Placement (2025)")
plt.xlabel("Time")
plt.ylabel("Cumulative Market Adoption (%)")
plt.ylim(0, 105)
plt.xlim(0, 20)
plt.legend()
plt.grid(alpha=0.3)

# Save for download
plt.tight_layout()

#plt.savefig("/mnt/data/genai_diffusion_consumer_vs_enterprise_2025_corrected.png", dpi=220)

plt.show()
```
