```python
import os
import textwrap
import matplotlib.pyplot as plt
import networkx as nx

# ============================================================
# Event-driven "graph database" representation (directed graph)
# ============================================================
# We model each Deal Flow as a directed property graph:
#   - Nodes: belief states (S0..S6, W, L) with properties (Pwin, label)
#   - Edges: actions/events with properties (seq, class, label)
# Cycles are represented explicitly as:
#   - self-loops (e.g., S6 -> S6)
#   - back-edges (e.g., S3 -> S2 regression)
#
# Output: one PNG per deal: deal_flow_01.png ... deal_flow_10.png
# Saved to: /Users/rudolphsurovcik/Library/CloudStorage/OneDrive-Personal/DAI Exhibit Processing/Markov
# ============================================================

OUT_DIR = "/Users/rudolphsurovcik/Library/CloudStorage/OneDrive-Personal/DAI Exhibit Processing/Markov"
os.makedirs(OUT_DIR, exist_ok=True)

# --- Belief-state mapping (node property) ---
PWIN = {
    "S0": 0.10, "S1": 0.22, "S2": 0.35, "S3": 0.45,
    "S4": 0.65, "S5": 0.90, "S6": 0.95,
    "W":  1.00, "L":  0.00
}

STATE_NAMES = {
    "S0": "Start",
    "S1": "Internal\nAlignment",
    "S2": "Exec\nAccess",
    "S3": "Technical\nValidation",
    "S4": "Value\nAnchored",
    "S5": "Exec\nCommitment",
    "S6": "Commercial\nLock",
    "W":  "Closed\nWon",
    "L":  "Closed\nLost"
}

# --- Action class tags (edge property) ---
EXEC = "Executive"
TECH = "Technical"
COMM = "Commercial"
STALL = "Stall/No-response"
TERM = "Terminal"

# --- 10 deal flows: each step specifies transition to a new state and action metadata ---
# Each tuple: (seq, from_state, to_state, action_label, action_class)
deal_edges = {
    1: [
        (1, "S0", "S1", "Partner strategy session", EXEC),
        (2, "S1", "S2", "Economic buyer intro meeting", EXEC),
        (3, "S2", "S3", "Architecture deep dive", TECH),
        (4, "S3", "S4", "Value reframe w/ sponsor", EXEC),
        (5, "S4", "S6", "Procurement kickoff + initial redlines", COMM),
        (6, "S6", "S6", "Discount request [recursive]", COMM),  # cycle (self-loop)
        (7, "S6", "W",  "Commercial alignment + signature", COMM),
    ],
    2: [
        (1, "S0", "S1", "SDR discovery + internal alignment", EXEC),
        (2, "S1", "S2", "Exec access meeting", EXEC),
        (3, "S2", "S1", "Pricing call too early [regression]", COMM),  # back-edge cycle potential
        (4, "S1", "S3", "Technical deep dive", TECH),
        (5, "S3", "S4", "Partner-led value reframe", EXEC),
        (6, "S4", "S5", "Sponsor commitment", EXEC),
        (7, "S5", "S6", "Legal/procurement finalization", COMM),
        (8, "S6", "W",  "Signature / close", COMM),
    ],
    3: [
        (1, "S0", "S1", "Partner strategy session", EXEC),
        (2, "S1", "S2", "Sponsor intro", EXEC),
        (3, "S2", "S3", "Security review", TECH),
        (4, "S3", "S3", "Remediation + follow-up security review [recursive]", TECH),  # self-loop
        (5, "S3", "S4", "Final security approval", TECH),
        (6, "S4", "S5", "Exec commitment", EXEC),
        (7, "S5", "S6", "Commercial lock", COMM),
        (8, "S6", "W",  "Close", COMM),
    ],
    4: [
        (1, "S0", "S0", "Outreach + discovery attempt [no lift]", EXEC),  # self-loop
        (2, "S0", "S0", "No response [recursive]", STALL),               # self-loop
        (3, "S0", "S1", "Re-engagement attempt w/ new angle", EXEC),
        (4, "S1", "S0", "Meeting cancelled / no-show [regression]", STALL),  # back edge
        (5, "S0", "L",  "Competitor selected / lost", TERM),
    ],
    5: [
        (1, "S0", "S1", "Partner alignment", EXEC),
        (2, "S1", "S2", "Exec access", EXEC),
        (3, "S2", "S3", "Technical validation", TECH),
        (4, "S3", "S3", "Proposal redraft [recursive]", COMM),           # self-loop
        (5, "S3", "S2", "Proposal redraft #2 [recursive regression]", COMM),  # back-edge
        (6, "S2", "S4", "Value reframe with sponsor", EXEC),
        (7, "S4", "S5", "Sponsor commitment", EXEC),
        (8, "S5", "S6", "Commercial lock", COMM),
        (9, "S6", "W",  "Close", COMM),
    ],
    6: [
        (1, "S0", "S1", "Partner alignment", EXEC),
        (2, "S1", "S2", "Sponsor meeting", EXEC),
        (3, "S2", "S4", "Value workshop + technical deep dive", TECH),
        (4, "S4", "S5", "Sponsor commitment", EXEC),
        (5, "S5", "S4", "Procurement demands + unfavorable terms [regression]", COMM),  # back-edge
        (6, "S4", "S4", "Discount escalation loop [recursive]", COMM),                 # self-loop
        (7, "S4", "L",  "No decision / budget reallocated", TERM),
    ],
    7: [
        (1, "S0", "S1", "Internal alignment", EXEC),
        (2, "S1", "S2", "Champion introduces exec sponsor", EXEC),
        (3, "S2", "S5", "Direct exec escalation by Partner", EXEC),  # jump
        (4, "S5", "S6", "Commercial lock", COMM),
        (5, "S6", "W",  "Close", COMM),
    ],
    8: [
        (1, "S0", "S1", "Partner strategy session", EXEC),
        (2, "S1", "S2", "Exec access meeting", EXEC),
        (3, "S2", "S3", "Competitive teardown workshop", TECH),
        (4, "S3", "S4", "Exec value reframe", EXEC),
        (5, "S4", "S4", "Competitor counteroffer → sponsor re-brief [recursive hold]", EXEC),  # self-loop
        (6, "S4", "S5", "Sponsor commitment", EXEC),
        (7, "S5", "S6", "Legal redlines", COMM),
        (8, "S6", "S6", "Last-minute redline loop [recursive]", COMM),  # self-loop
        (9, "S6", "W",  "Close", COMM),
    ],
    9: [
        (1, "S0", "S1", "Internal alignment", EXEC),
        (2, "S1", "S2", "Exec access", EXEC),
        (3, "S2", "S3", "Technical validation", TECH),
        (4, "S3", "S3", "Scope expansion request [recursive]", COMM),   # self-loop
        (5, "S3", "S4", "Proposal redraft for new scope [value clarified]", COMM),
        (6, "S4", "S4", "Another scope expansion request [recursive hold]", COMM),  # self-loop
        (7, "S4", "S5", "Exec sponsor alignment on phased rollout", EXEC),
        (8, "S5", "S6", "Commercial lock", COMM),
        (9, "S6", "W",  "Close", COMM),
    ],
    10: [
        (1, "S0", "S1", "Partner alignment", EXEC),
        (2, "S1", "S2", "Exec access", EXEC),
        (3, "S2", "S2", "Technical deep dive reveals gap [no lift]", TECH),      # self-loop
        (4, "S2", "S2", "Remediation attempt + re-test [recursive]", TECH),      # self-loop
        (5, "S2", "L",  "Buyer selects alternative", TERM),
    ],
}

def _wrap(s: str, width: int = 26) -> str:
    return "\n".join(textwrap.wrap(s, width=width))

def build_deal_graph(flow_num: int) -> nx.DiGraph:
    G = nx.DiGraph(flow_id=flow_num)

    # Add all belief-state nodes with properties
    for nid, p in PWIN.items():
        G.add_node(
            nid,
            pwin=p,
            label=f"{nid}\n{STATE_NAMES.get(nid, nid)}\nP={p:.2f}"
        )

    # Add edges for this deal with properties
    for seq, src, dst, action, aclass in deal_edges[flow_num]:
        G.add_edge(
            src, dst,
            seq=seq,
            action=action,
            action_class=aclass,
            label=f"({seq}) {_wrap(action, 26)}"
        )

    return G

def plot_deal_graph(flow_num: int, out_dir: str) -> str:
    G = build_deal_graph(flow_num)

    # Layout: spring handles cycles reasonably without manual placement
    pos = nx.spring_layout(G, seed=flow_num, k=1.0, iterations=250)

    # Draw nodes with labels
    plt.figure(figsize=(12, 8))
    nx.draw_networkx_nodes(G, pos, node_size=2200)
    nx.draw_networkx_labels(
        G, pos,
        labels={n: G.nodes[n]["label"] for n in G.nodes},
        font_size=8
    )

    # Draw edges + arrows
    nx.draw_networkx_edges(G, pos, arrows=True, arrowstyle="-|>", arrowsize=18, width=1.6)

    # Edge labels show sequence + action (properties)
    edge_labels = {(u, v): G.edges[u, v]["label"] for u, v in G.edges}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7, label_pos=0.5)

    plt.title(f"Deal Flow #{flow_num} — Event-Driven Property Graph (cycles explicit)")
    plt.axis("off")
    plt.tight_layout()

    out_path = os.path.join(out_dir, f"deal_flow_{flow_num:02d}.png")
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()
    return out_path

# --- Generate and save all 10 deal graphs ---
saved = []
for n in range(1, 11):
    saved.append(plot_deal_graph(n, OUT_DIR))

print("Saved files:")
for p in saved:
    print(p)

#What this gives you (conceptually “graph database”)
#	•	Each plot is a directed property graph (nodes + edge properties), with cycles explicitly represented as self-loops and back-edges.
#	•	Every edge is labeled with its sequence number and action, so the graph is still “event-order readable” even though it’s not a linear chart.
#	•	Each PNG is saved as:
#	•	deal_flow_01.png … deal_flow_10.png
#	•	in your OneDrive folder path.

#If you want the “graph database” aspect to be even more literal (e.g., export these into Neo4j Cypher CREATE statements or a GraphML file per deal), say so and I’ll output that too.
```
