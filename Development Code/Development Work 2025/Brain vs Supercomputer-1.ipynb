```python
# genome_to_cognition.py

stages = [
    {
        "title": "DNA Sequence",
        "info": "~0.8 GB of raw data",
        "details": [
            "3.2 billion base pairs",
            "4 letters: A, T, C, G",
            "Encodes ~20,000 genes",
        ],
        "tagline": "Compressed source code of life",
    },
    {
        "title": "Epigenetic & Cellular Regulation",
        "info": "~100–1,000× data expansion",
        "details": [
            "Gene expression patterns",
            "Methylation & histone modification",
            "Cell–cell signaling networks",
        ],
        "tagline": "Dynamic runtime control of the code",
    },
    {
        "title": "Neural Architecture & Cognition",
        "info": "~2.5 PB (petabytes)",
        "details": [
            "~86 billion neurons",
            "~100 trillion synapses",
            "Plastic, self-modifying, experience-driven",
        ],
        "tagline": "Emergent intelligence, reasoning, and self-awareness",
    },
]


def print_stages():
    """Pretty-print the stages in text form."""
    print("From Genome to Cognition: The Expansion of Information\n")
    for i, stage in enumerate(stages, start=1):
        print(f"{i}. {stage['title']}")
        print(f"   {stage['info']}")
        for d in stage["details"]:
            print(f"   - {d}")
        print(f"   >> {stage['tagline']}\n")


if __name__ == "__main__":
    print_stages()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
from matplotlib.patches import FancyArrowPatch

def plot_stages():
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.set_axis_off()

    # x positions for three boxes
    xs = [0.1, 0.6, 1.0]
    y = 0.5
    width = 0.35
    height = 0.35

    for x, stage in zip(xs, stages):
        # Draw box
        box = FancyBboxPatch(
            (x, y - height / 2),
            width,
            height,
            boxstyle="round,pad=0.02",
            edgecolor="black",
            facecolor="white"
        )
        ax.add_patch(box)

        # Add text
        text = (
            f"{stage['title']}\n\n"
            f"{stage['info']}\n\n" +
            "\n".join(stage["details"]) +
            f"\n\n{stage['tagline']}"
        )
        ax.text(
            x + width / 2,
            y,
            text,
            ha="center",
            va="center",
            wrap=True,
            fontsize=9,
        )

    # Arrows between boxes
    for i in range(len(xs) - 1):
        arrow = FancyArrowPatch(
            (xs[i] + width, y),
            (xs[i + 1], y),
            arrowstyle="->",
            mutation_scale=15,
        )
        ax.add_patch(arrow)

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print_stages()
    # Uncomment to show the figure:
    plot_stages()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------
# Approximate comparison numbers (order-of-magnitude)
# ---------------------------------------------------------

systems = ["Human Brain", "Frontier\nSupercomputer"]

# 1) Compute throughput (FLOPS)
# Brain: ~1e16 FLOPS (10 PFLOPS)
# Frontier: ~1.7e18 FLOPS (1.7 EFLOPS)
flops = np.array([1e16, 1.7e18])

# 2) Memory / storage capacity (Petabytes)
# Brain: ~2.5 PB (functional synaptic capacity analog)
# Frontier: ~9.2 PB RAM (published ballpark)
memory_pb = np.array([2.5, 9.2])

# 3) Power consumption (Watts)
# Brain: ~20 W
# Frontier: ~20e6 W (20 MW)
power_w = np.array([20.0, 20e6])

# ---------------------------------------------------------
# Create figure with three subplots
# ---------------------------------------------------------

fig, axes = plt.subplots(3, 1, figsize=(10, 12))
plt.subplots_adjust(hspace=0.4)

# ========== Panel 1: FLOPS (log scale) ==========
ax = axes[0]
bars = ax.bar(systems, flops)

ax.set_yscale("log")
ax.set_ylabel("Compute Throughput (FLOPS, log scale)")
ax.set_title("Human Brain vs Frontier Supercomputer\nApproximate Compute Throughput")

# Add labels
flops_labels = ["~1×10¹⁶ FLOPS", "~1.7×10¹⁸ FLOPS"]
for bar, label in zip(bars, flops_labels):
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
    )

# ========== Panel 2: Memory (Petabytes) ==========
ax = axes[1]
bars = ax.bar(systems, memory_pb)

ax.set_ylabel("Memory / Capacity (PB)")
ax.set_title("Approximate Memory / Capacity")

mem_labels = ["~2.5 PB\n(unified synaptic capacity)",
              "~9.2 PB RAM"]
for bar, label in zip(bars, mem_labels):
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
    )

# ========== Panel 3: Power (Watts, log scale) ==========
ax = axes[2]
bars = ax.bar(systems, power_w)

ax.set_yscale("log")
ax.set_ylabel("Power (Watts, log scale)")
ax.set_title("Power Consumption")

power_labels = ["~20 W", "~20,000,000 W\n(20 MW)"]
for bar, label in zip(bars, power_labels):
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
    )

# Common X tick formatting
for ax in axes:
    ax.tick_params(axis='x', labelrotation=0)

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# Optional: Save to high-resolution PNG
# ---------------------------------------------------------
# fig.savefig("brain_vs_frontier.png", dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------
# Parameters / Assumptions
# ---------------------------------------------------------

systems = ["Human Brain", "Frontier Supercomputer"]

# Approximate compute throughput (very rough, order-of-magnitude)
# Brain: ~1e15 ops/s (1 PFLOP-ish)
# Frontier: ~1.7e18 FLOPS (1.7 EFLOPS)
ops_per_sec = np.array([1e15, 1.7e18])

# Power (Watts)
brain_power_w = 20            # 20 W
frontier_power_w = 20_000_000 # 20 MW
power_w = np.array([brain_power_w, frontier_power_w])

# Time window: 30 minutes
time_hours = 0.5           # hours
time_seconds = 30 * 60     # 1800 seconds

# Energy in Joules (P * t)
brain_energy_j = brain_power_w * time_seconds
frontier_energy_j = frontier_power_w * time_seconds

# Convert to kWh for easier intuition: 1 kWh = 3.6e6 J
J_PER_KWH = 3.6e6
brain_energy_kwh = brain_energy_j / J_PER_KWH
frontier_energy_kwh = frontier_energy_j / J_PER_KWH

energy_kwh = np.array([brain_energy_kwh, frontier_energy_kwh])

# ---------------------------------------------------------
# Plotting
# ---------------------------------------------------------

fig, axes = plt.subplots(3, 1, figsize=(10, 12))
plt.subplots_adjust(hspace=0.4)

# ---------- Panel 1: Compute Throughput (ops/s, log scale) ----------
ax1 = axes[0]
bars1 = ax1.bar(systems, ops_per_sec, color=["purple", "lightcoral"])

ax1.set_yscale("log")
ax1.set_ylabel("Compute Throughput (ops/s, log scale)")
ax1.set_title("Approximate Compute Throughput")

ops_labels = ["~1×10¹⁵ ops/s", "~1.7×10¹⁸ ops/s"]
for bar, label in zip(bars1, ops_labels):
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
    )

# ---------- Panel 2: Power (Watts, log scale) ----------
ax2 = axes[1]
bars2 = ax2.bar(systems, power_w, color=["purple", "lightcoral"])

ax2.set_yscale("log")
ax2.set_ylabel("Power (Watts, log scale)")
ax2.set_title("Power Draw")

power_labels = [f"{brain_power_w} W", f"{frontier_power_w:,} W"]
for bar, label in zip(bars2, power_labels):
    height = bar.get_height()
    ax2.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
    )

# ---------- Panel 3: Energy over 30 Minutes (kWh, log scale) ----------
ax3 = axes[2]
bars3 = ax3.bar(systems, energy_kwh, color=["purple", "lightcoral"])

ax3.set_yscale("log")
ax3.set_ylabel("Energy over 30 minutes (kWh, log scale)")
ax3.set_title("Energy Consumed in 30 Minutes of Operation")

energy_labels = [f"{brain_energy_kwh:.2f} kWh", f"{frontier_energy_kwh:,.0f} kWh"]
for bar, label in zip(bars3, energy_labels):
    height = bar.get_height()
    ax3.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
    )

for ax in axes:
    ax.tick_params(axis="x", labelrotation=0)

plt.tight_layout()
plt.show()

# Optional: Save as high-res PNG
# fig.savefig("brain_vs_frontier_30min_energy.png", dpi=300)
```
