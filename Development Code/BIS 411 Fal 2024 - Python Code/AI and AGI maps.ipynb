```python
import textwrap

import matplotlib.pyplot as plt

# Data
years = [2020, 2025, 2028, 2030, 2035, 2040, 2050]
milestone_index = [0, 1, 2, 3, 4, 5, 6]
milestones = [
    "GPT-2 released; expert interest begins",
    "94% of enterprises using AI; only 14% aiming for enterprise-wide AI by 2025",
    "Consumer-facing generative AI tools emerge; enterprise interest surges",
    "Target year for enterprise-wide AI adoption",
    "Generative AI accelerates workforce automation by a decade",
    "Projected start of 50% work automation",
    "Projected completion of 50% work automation"
]

# Plot setup
plt.figure(figsize=(12, 6))
plt.plot(years, milestone_index, marker='o', color='royalblue', linewidth=2)

# Annotate milestones
for x, y, label in zip(years, milestone_index, milestones):
    wrapped_label = "\n".join(textwrap.wrap(label, width=35))
    plt.text(x, y + 0.2, wrapped_label, fontsize=9, ha='center', va='bottom')

# Economic impact annotation
plt.text(2047, 1, "$2.6–4.4 trillion annual value\n7% global GDP boost", 
         fontsize=10, color='green', ha='right', va='bottom')

# Axis labels and title
plt.title("Enterprise Adoption Curve of Generative AI\nSource: MIT Technology Review Insights", fontsize=14, weight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Milestone Index", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Display
plt.tight_layout()
plt.show()
# Save the plot as a PNG file
# plt.savefig("enterprise_adoption_curve.png", format='png')

# Set the output path to "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"
output_path = "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"

#save this file to the output path

plt.savefig(f"{output_path}/enterprise_adoption_curve.png",format='png')
```

```python
import matplotlib.pyplot as plt

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Positions for nodes
positions = {
    "User": (0.1, 0.5),
    "Agentic Chatbot": (0.4, 0.5),
    "Multi-Agent System": (0.7, 0.5),
    "Agent A": (0.9, 0.75),
    "Agent B": (0.9, 0.6),
    "Agent C": (0.9, 0.45),
    "Agent D": (0.9, 0.3)
}

# Draw nodes
ax.text(*positions["User"], "👤 User", fontsize=12, ha="center", va="center", bbox=dict(boxstyle="round,pad=0.3", fc="lightblue"))
ax.text(*positions["Agentic Chatbot"], "🤖 Agentic\nChatbot", fontsize=12, ha="center", va="center", bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen"))
ax.text(*positions["Multi-Agent System"], "🤖 Multi-Agent\nSystem", fontsize=12, ha="center", va="center", bbox=dict(boxstyle="round,pad=0.3", fc="khaki"))

# Multi-agent smaller nodes
for agent in ["Agent A", "Agent B", "Agent C", "Agent D"]:
    ax.text(*positions[agent], agent, fontsize=10, ha="center", va="center", bbox=dict(boxstyle="round,pad=0.3", fc="wheat"))

# Draw arrows for single chatbot
ax.annotate("", xy=positions["Agentic Chatbot"], xytext=positions["User"], arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=positions["User"], xytext=positions["Agentic Chatbot"], arrowprops=dict(arrowstyle="->", lw=2))

# Draw arrows for multi-agent system
ax.annotate("", xy=positions["Multi-Agent System"], xytext=positions["User"], arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=positions["User"], xytext=positions["Multi-Agent System"], arrowprops=dict(arrowstyle="->", lw=2))

# Connect multi-agent system to individual agents
for agent in ["Agent A", "Agent B", "Agent C", "Agent D"]:
    ax.annotate("", xy=positions[agent], xytext=positions["Multi-Agent System"], arrowprops=dict(arrowstyle="->", lw=1))

# Formatting
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")
ax.set_title("Agentic AI Chatbot vs Multi-Agent System", fontsize=14, weight="bold")

plt.show()

# Set the output path to "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"
output_path = "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"

#save this file to the output path

plt.savefig(f"{output_path}/MultiAgent Systems.png",format='png')
```

```python
import matplotlib.pyplot as plt

# --- Data (illustrative estimates) ---
years = [1990, 2000, 2010, 2020, 2025, 2030]
adoption = [5, 10, 15, 25, 45, 75]  # Estimated percentage adoption

# --- Create the chart ---
plt.figure(figsize=(10, 6))
plt.plot(years, adoption, marker='o', linewidth=2)

# Milestone labels
milestones = {
    1990: "Early MAS research\n(simulations, robotics)",
    2000: "Agent-based modeling in\nacademia & niche industries",
    2010: "Wider use in\nlogistics & traffic systems",
    2020: "LLM rise; agent frameworks emerge",
    2025: "Enterprise pilots:\nfinance, healthcare, supply chain",
    2030: "Mainstream enterprise\nMAS adoption (predicted)"
}

for year, label in milestones.items():
    idx = years.index(year)
    plt.text(year, adoption[idx] + 3, label, ha="center", va="bottom",
             fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow"))

# --- Formatting ---
plt.title("Timeline of Multi-Agent System (MAS) Adoption", fontsize=14, weight="bold")
plt.xlabel("Year")
plt.ylabel("Estimated Adoption (%)")
plt.ylim(0, 100)
plt.grid(True, linestyle="--", alpha=0.6)

# --- Save a copy ---
outfile = "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data/MultiAgent Systems Timeline.png"
plt.savefig(outfile, dpi=200, bbox_inches="tight")

plt.show()

# Set the output path to "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"
output_path = "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"

#save this file to the output path

#plt.savefig(f"{output_path}/MultiAgent Systems Timeline.png",format='png')
```
