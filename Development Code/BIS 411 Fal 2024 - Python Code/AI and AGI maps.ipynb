```python
import matplotlib.pyplot as plt
import textwrap

# Stages of AI progression
stages = ["Narrow AI", "Proto-AGI", "AGI"]
capabilities = [20, 50, 100]  # illustrative "capability index"

# Plot progression
plt.figure(figsize=(10, 6))
plt.plot(stages, capabilities, marker="o", linewidth=3, color="teal")

# Annotate descriptions with wrapped text
labels = {
    "Narrow AI": "Specialized, single-task (Spam filter, recommender systems)",
    "Proto-AGI": "Early generalization, tool use, planning, multi-modal reasoning (Current LLM agents)",
    "AGI": "Human-level general intelligence: Adaptive, robust, creative"
}

for stage, cap in zip(stages, capabilities):
    wrapped = "\n".join(textwrap.wrap(labels[stage], width=35))
    plt.text(stage, cap - 10, wrapped, ha="center", fontsize=9,
             bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow"))

# Formatting
plt.title("Progression from Narrow AI → Proto-AGI → AGI", fontsize=14, weight="bold")
plt.ylabel("General Capability (illustrative index)")
plt.ylim(0, 120)
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt

# --- Timeline events (arc of AI) ---
years = [1956, 1980, 1997, 2012, 2022, 2025]
labels = [
    "1956\nSymbolic AI\n(Dartmouth: McCarthy, Minsky, Simon)",
    "1980\nMachine Learning\n(Expert systems, Backpropagation)",
    "1997\nBig Data + Early ML\n(IBM Deep Blue, stats learning)",
    "2012\nDeep Learning\n(AlexNet: Hinton, Bengio, LeCun)",
    "2022\nGenerative AI\n(ChatGPT, mainstream LLMs)",
    "2025\nProto-AGI\n(Multi‑modal, agentic systems)"
]

# --- AI Winters (approximate ranges) ---
ai_winters = [
    (1974, 1980, "First AI Winter"),
    (1987, 1993, "Second AI Winter"),
]

# --- Build the figure ---
fig, ax = plt.subplots(figsize=(14, 5))

# Base horizontal timeline
xmin, xmax = min(years) - 5, max(years) + 5
ax.hlines(1, xmin, xmax, color="gray", linewidth=1)

# Shade AI Winters
for start, end, name in ai_winters:
    ax.axvspan(start, end, ymin=0.63, ymax=0.77, color="lightcoral", alpha=0.3)
    ax.text((start + end) / 2, 0.3, name, ha="center", va="center",
            fontsize=9, weight="bold", color="darkred")

# Plot event markers
ax.scatter(years, [1]*len(years), color="teal", s=100, zorder=3)

# Place event labels alternating above/below the line
for i, (yr, text) in enumerate(zip(years, labels)):
    offset = 0.28 if i % 2 == 0 else -0.28
    ax.text(yr, 1 + offset, text, ha="center", va="center",
            fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow"))

# Formatting
ax.set_xlim(xmin, xmax)
ax.set_ylim(0.2, 1.7)
ax.axis("off")
ax.set_title(
    "Timeline of AI Development with AI Winters\n"
    "Symbolic AI → Machine Learning → Deep Learning → Generative AI → Proto‑AGI",
    fontsize=12, weight="bold"
)

plt.show()

# Save a copy (optional)
# Set the output path to "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"
output_path = "/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2025/BIS 411 - Fall 2025/Tufte Data"

#save this file to the output path

plt.savefig(f"{output_path}/Timeline.png",format='png')
```

```python
import matplotlib.pyplot as plt
%pip install networkx
import networkx as nx
import matplotlib.patches as mpatches

# Define the PayPal Mafia members and companies
members = [
    "Elon Musk", "Peter Thiel", "Max Levchin", "Reid Hoffman",
    "David Sacks", "Steve Chen", "Chad Hurley", "Jawed Karim",
    "Jeremy Stoppelman", "Roelof Botha", "Keith Rabois"
]

# Companies grouped by industry
companies_by_industry = {
    "Transportation/Space": ["Tesla", "SpaceX", "The Boring Company"],
    "AI/Neurotech": ["Neuralink", "xAI", "Inflection AI", "Palantir"],
    "Fintech": ["Affirm", "Slide", "Square (Block)"],
    "Venture Capital": ["Founders Fund", "Greylock Partners", "Craft Ventures", "Sequoia Capital"],
    "Social Media/Networking": ["LinkedIn", "Facebook", "Yammer"],
    "Video/Media": ["YouTube", "MixBit"],
    "Local/Reviews": ["Yelp"],
    "Delivery/Marketplace": ["DoorDash", "Airbnb"]
}

# Flatten companies list
companies = [c for clist in companies_by_industry.values() for c in clist]

# Relationships (edges) with role type (founded vs funded)
edges = [
    ("Elon Musk", "Tesla", "founded"), ("Elon Musk", "SpaceX", "founded"), ("Elon Musk", "Neuralink", "founded"),
    ("Elon Musk", "The Boring Company", "founded"), ("Elon Musk", "xAI", "founded"),
    ("Peter Thiel", "Palantir", "founded"), ("Peter Thiel", "Founders Fund", "founded"),
    ("Max Levchin", "Affirm", "founded"), ("Max Levchin", "Slide", "founded"),
    ("Reid Hoffman", "LinkedIn", "founded"), ("Reid Hoffman", "Greylock Partners", "funded"), ("Reid Hoffman", "Inflection AI", "founded"),
    ("David Sacks", "Yammer", "founded"), ("David Sacks", "Craft Ventures", "founded"),
    ("Steve Chen", "YouTube", "founded"), ("Chad Hurley", "YouTube", "founded"), ("Jawed Karim", "YouTube", "founded"),
    ("Chad Hurley", "MixBit", "founded"),
    ("Jeremy Stoppelman", "Yelp", "founded"),
    ("Roelof Botha", "Sequoia Capital", "funded"),
    ("Keith Rabois", "DoorDash", "funded"), ("Keith Rabois", "Airbnb", "funded"),
    ("Keith Rabois", "Facebook", "funded"), ("Keith Rabois", "Square (Block)", "funded")
]

# Create bipartite graph
G = nx.DiGraph()
G.add_nodes_from(members, bipartite=0)
G.add_nodes_from(companies, bipartite=1)
for m, c, role in edges:
    if role == "founded":
        G.add_edge(m, c, role="founded")
    else:
        G.add_edge(m, c, role="funded")

# Assign colors by category
industry_colors = {
    "Transportation/Space": "lightcoral",
    "AI/Neurotech": "lightyellow",
    "Fintech": "lightgreen",
    "Venture Capital": "lightpink",
    "Social Media/Networking": "lightskyblue",
    "Video/Media": "plum",
    "Local/Reviews": "khaki",
    "Delivery/Marketplace": "orange"
}

node_colors = []
for node in G.nodes():
    if node in members:
        node_colors.append("lightblue")
    else:
        color = "lightgray"
        for industry, comps in companies_by_industry.items():
            if node in comps:
                color = industry_colors[industry]
                break
        node_colors.append(color)

# Layout for bipartite graph (spread out more vertically)
pos = {}
pos.update((node, (0, i * 2)) for i, node in enumerate(members))  # Left column: members
pos.update((node, (1, i * 1.5)) for i, node in enumerate(companies))  # Right column: companies

# Draw graph
plt.figure(figsize=(26, 22))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=5000, node_color=node_colors)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight="bold")

# Draw edges with different styles
founded_edges = [(u, v) for u, v, d in G.edges(data=True) if d['role'] == "founded"]
funded_edges = [(u, v) for u, v, d in G.edges(data=True) if d['role'] == "funded"]

nx.draw_networkx_edges(G, pos, edgelist=founded_edges, edge_color="black", arrows=True, width=2)
nx.draw_networkx_edges(G, pos, edgelist=funded_edges, edge_color="gray", style="dashed", arrows=True, width=2)

# Legend
legend_patches = [mpatches.Patch(color="lightblue", label="PayPal Mafia Members")]
for industry, color in industry_colors.items():
    legend_patches.append(mpatches.Patch(color=color, label=industry))

edge_legend = [
    mpatches.Patch(color="black", label="Founded (solid)"),
    mpatches.Patch(color="gray", label="Funded (dashed)")
]

plt.legend(handles=legend_patches + edge_legend, loc="upper left", bbox_to_anchor=(1, 1), fontsize=12)

plt.title("The PayPal Mafia and the Companies They Founded or Funded\n(Solid = Founded, Dashed = Funded)", fontsize=18, weight="bold")
plt.axis("off")
plt.show()
```
