```python
# Fixing the mismatch in data length for the radar chart
# Re-define the categories and values correctly

import numpy as np
import matplotlib.pyplot as plt

categories = [
    "Emotional Expressiveness",
    "Verbal Affection",
    "Physical Affection",
    "Conflict Expression",
    "Emotional Privacy",
    "Public Emotional Display"
]

# Values for each category on a 010 scale
italy_values = [9, 9, 8, 8, 3, 9]
slovakia_values = [5, 4, 5, 3, 8, 4]

# Make the radar chart circular
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]
italy_values += italy_values[:1]
slovakia_values += slovakia_values[:1]

# Create the radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

ax.plot(angles, italy_values, linewidth=2, linestyle='solid', label='Italy')
ax.fill(angles, italy_values, alpha=0.25)

ax.plot(angles, slovakia_values, linewidth=2, linestyle='solid', label='Slovakia')
ax.fill(angles, slovakia_values, alpha=0.25)

# Add labels and title
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), categories)
ax.set_title("Emotional Expression in Families: Italy vs Slovakia", size=16, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.tight_layout()
plt.show()
```

```python
# Re-import necessary libraries after code execution environment reset
import plotly.express as px
import pandas as pd

# Data for the cultural influence of Saints Cyril and Methodius
data = {
    "Country": [
        "Slovakia", "Czechia", "North Macedonia", "Bulgaria", "Serbia",
        "Russia", "Ukraine", "Belarus", "Croatia", "Slovenia",
        "Bosnia and Herzegovina", "Montenegro", "Poland", "Romania",
        "Greece", "Italy"
    ],
    "Influence": [
        "Core Legacy", "Core Legacy", "Core Legacy", "Core Legacy", "Core Legacy",
        "Cultural Influence", "Cultural Influence", "Cultural Influence", "Cultural Influence", "Cultural Influence",
        "Cultural Influence", "Cultural Influence", "Limited Influence", "Limited Influence",
        "Byzantine Origin", "Papal Support"
    ]
}

df = pd.DataFrame(data)

# Create choropleth map
fig = px.choropleth(
    df,
    locations="Country",
    locationmode="country names",
    color="Influence",
    title="Cultural Legacy of Saints Cyril and Methodius",
    color_discrete_map={
        "Core Legacy": "#1f77b4",
        "Cultural Influence": "#ff7f0e",
        "Limited Influence": "#2ca02c",
        "Byzantine Origin": "#d62728",
        "Papal Support": "#9467bd"
    },
    projection="natural earth"
)

fig.update_layout(
    legend_title_text='Cultural-Historical Influence',
    title_x=0.5
)

fig.show()
```

```python
# Re-import necessary libraries after code execution state reset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Generate sample data
np.random.seed(42)
right_skew = np.random.exponential(scale=2, size=1000)
left_skew = -1 * np.random.exponential(scale=2, size=1000) + 10  # shift to make data positive

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Right-skewed distribution
sns.histplot(right_skew, kde=True, ax=axes[0], color='skyblue', bins=30)
axes[0].set_title('Right-Skewed Distribution')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Left-skewed distribution
sns.histplot(left_skew, kde=True, ax=axes[1], color='salmon', bins=30)
axes[1].set_title('Left-Skewed Distribution')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Recalculate mean and median for both distributions
right_mean = np.mean(right_skew)
right_median = np.median(right_skew)
left_mean = np.mean(left_skew)
left_median = np.median(left_skew)

# Create subplots again with overlays
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Right-skewed distribution
sns.histplot(right_skew, kde=True, ax=axes[0], color='skyblue', bins=30)
axes[0].axvline(right_mean, color='blue', linestyle='--', linewidth=2, label=f'Mean: {right_mean:.2f}')
axes[0].axvline(right_median, color='green', linestyle='--', linewidth=2, label=f'Median: {right_median:.2f}')
axes[0].set_title('Right-Skewed Distribution')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Left-skewed distribution
sns.histplot(left_skew, kde=True, ax=axes[1], color='salmon', bins=30)
axes[1].axvline(left_mean, color='blue', linestyle='--', linewidth=2, label=f'Mean: {left_mean:.2f}')
axes[1].axvline(left_median, color='green', linestyle='--', linewidth=2, label=f'Median: {left_median:.2f}')
axes[1].set_title('Left-Skewed Distribution')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].legend()

plt.tight_layout()
plt.show()
```
