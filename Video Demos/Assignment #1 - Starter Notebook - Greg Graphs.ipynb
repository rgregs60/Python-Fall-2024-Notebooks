<p>**Create a Bar Graph Using Product Category Show Total Revenue**<p>
</div>

```python
import seaborn as sns   # importing seaborn library
import matplotlib.pyplot as plt  # importing matplotlib library
import matplotlib.ticker as mtick  # importing ticker for formatting the y-axis
import pandas as pd  # importing pandas library
from statsmodels.nonparametric.smoothers_lowess import lowess  # importing lowess for smoothing

# Load the dataset
Assign_1 = pd.read_excel('BIS411_grs3_lehigh_edu_Assignment1.xlsx') 

# Summarize revenue by product category
revenue_by_category = Assign_1.groupby('Product_Category')['Revenue'].sum().reset_index()

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the sum of revenue by product category
plt.bar(revenue_by_category['Product_Category'], revenue_by_category['Revenue'], color='#800000')

# Set the face color of the axes to white
plt.gca().set_facecolor('white')  
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M'))

# Adding labels and title
plt.title('Product Category vs Revenue (2018 to 2023)', fontsize=14)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Revenue (in millions)', fontsize=12)
plt.gca().yaxis.set_ticks_position('none')
plt.gca().set_ylim([0, 14000000])


# Set x-axis labels to be horizontal
plt.xticks(rotation=0, fontsize=11)

# turn off the border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# add data labels to the bars 
for i in range(len(revenue_by_category)):
    plt.text(i, revenue_by_category['Revenue'][i], f"${revenue_by_category['Revenue'][i]:,.0f}", ha='center', va='bottom', fontsize=10)



# Turn off the grid
plt.grid(False)

# Display the chart
plt.show()
```

<p>**Create a Line Graph for HVAC Showing Total Revenue by Year**<p>
</div>

```python
import seaborn as sns   # importing seaborn library
import matplotlib.pyplot as plt  # importing matplotlib library
import matplotlib.ticker as mtick  # importing ticker for formatting the y-axis
import pandas as pd  # importing pandas library
from statsmodels.nonparametric.smoothers_lowess import lowess  # importing lowess for smoothing

# Load the data set
Assign_1 = pd.read_excel('BIS411_grs3_lehigh_edu_Assignment1.xlsx')

# Filter the data for the Product_Category HVAC
Assign_1 = Assign_1[Assign_1['Product_Category'] == 'HVAC']

# Summarize revenue by year
revenue_by_year = Assign_1.groupby('Year')['Revenue'].sum().reset_index()

# Apply lowess smoothing
smoothed = lowess(revenue_by_year['Revenue'], revenue_by_year['Year'], frac=0.3)

# Set the style of the graph
plt.figure(figsize=(10, 6))

# Plotting the smoothed line graph
plt.plot(smoothed[:, 0], smoothed[:, 1], color='#000000', label='Smoothed Line')

# Plotting the regression line
sns.regplot(data=revenue_by_year, x='Year', y='Revenue', scatter=False, color='skyblue', label='Regression Line')

# Set the face color of the axes to white
plt.gca().set_facecolor('white')

# Adding labels and title
plt.title('Revenue by Year for HVAC', fontsize=14, color='#000000')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.gca().yaxis.set_ticks_position('none')

# Format the y-axis to display values in millions
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M'))

# Turn off the grid
plt.grid(False)

# turn off the border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# Adding legend
plt.legend()
plt.legend(loc='lower right')


# Display the chart
plt.show()
```

```python
import seaborn as sns   # importing seaborn library
import matplotlib.pyplot as plt  # importing matplotlib library
import matplotlib.ticker as mtick  # importing ticker for formatting the y-axis
import pandas as pd  # importing pandas library
from statsmodels.nonparametric.smoothers_lowess import lowess  # importing lowess for smoothing

# Load the data set
Assign_1 = pd.read_excel('BIS411_grs3_lehigh_edu_Assignment1.xlsx')

# Filter the data for the Product_Category HVAC
Assign_1 = Assign_1[Assign_1['Product_Category'] == 'HVAC']

# Summarize revenue by year
revenue_by_year = Assign_1.groupby('Year')['Revenue'].sum().reset_index()

# Apply lowess smoothing
smoothed = lowess(revenue_by_year['Revenue'], revenue_by_year['Year'], frac=0.6)

# Set the style of the graph
plt.figure(figsize=(10, 6))

# Plotting the smoothed line graph
plt.plot(smoothed[:, 0], smoothed[:, 1], color='#000000', label='Smoothed Line')

# Plotting the regression line
# sns.regplot(data=revenue_by_year, x='Year', y='Revenue', scatter=False, color='skyblue', label='Regression Line')

sns.regplot(data=revenue_by_year, x='Year', y='Revenue', scatter=False, color='skyblue', label='Regression Line', order=2)  # Add order=2 to fit a quadratic regression line


# Set the face color of the axes to white
plt.gca().set_facecolor('white')

# Adding labels and title
plt.title('Revenue by Year for HVAC', fontsize=14, color='#000000')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.gca().yaxis.set_ticks_position('none')

# Format the y-axis to display values in millions
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M'))

# Turn off the grid
plt.grid(False)

# turn off the border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# Adding legend
plt.legend()
plt.legend(loc='lower right')


# Display the chart
plt.show()
```
