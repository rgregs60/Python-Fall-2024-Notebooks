<h3>Data Vectors</h3>
<b>Basic Data Operations</b>
</div>

```python
###
### Example of creating a data frame in Python
###
%reset -f


import pandas as pd

# Create vectors for the data frame
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
salaries = [50000, 55000, 60000]

# Join vectors into a data frame using the DataFrame command
df = pd.DataFrame({'Name': names, 'Age': ages, 'Salary': salaries})

# View the data frame (In Jupyter, you can display the dataframe directly)
print(df)

# Print the data frame
print(df)

# The data frame is a table with rows and columns
print(df['Name']) # Print the Name column
print(df['Age']) # Print the Age column
print(df['Salary']) # Print the Salary column
```

```python
###
### Printing the Data Frame by Rows and Columns Using Indexing
###

import pandas as pd

# Printing the Data Frame by Rows and Columns Using Indexing 
print(df.iloc[0, :]) # Print the first row
print(df.iloc[1, :]) # Print the second row
print(df.iloc[2, :]) # Print the third row

print(df.iloc[:, 0]) # Print the first column
print(df.iloc[:, 1]) # Print the second column
print(df.iloc[:, 2]) # Print the third column

print(df.iloc[0, 0]) # Print the first row, first column
print(df.iloc[0, 1]) # Print the first row, second column
print(df.iloc[0, 2]) # Print the first row, third column

print(df.iloc[0:2, :]) # Print the first two rows
print(df.iloc[1:3, :]) # Print the second and third rows
print(df.iloc[0:2, 0:2]) # Print the first two rows and the first two columns

# Summary Operations using a Data frame
print(df['Salary'].sum()) 
print(df['Salary'].mean()) 
print(df['Salary'].median()) 
print(df['Salary'].std())
```

```python
###
### Read CSV File into a Data Frame and Clean Outliers - Part 1
###

%reset -f


import pandas as pd
import os

# Read the CSV file into a DataFrame

display(os.getcwd())
datapath = '/Users/rudolphsurovcik/Library/CloudStorage/GoogleDrive-grs3@lehigh.edu/My Drive/Lehigh Fall 2024/BIS 411 - Fall 2024/Python-Fall-2024-Notebooks/Data Files for Python'

os.chdir(datapath)  # set the working directory to the path where the data files are located 

bank_messy = pd.read_csv("bank_messy.csv")

# View the DataFrame (In Jupyter, you can display the dataframe directly)
print(bank_messy)

# Summary of the DataFrame
print(bank_messy.describe(include='all'))

# Identify outliers in the DataFrame bank_messy

# 1. Find row numbers for NAs in Income, Family, CCAvg, CreditCard, and Loan
print(bank_messy[bank_messy['Income'].isna()])
print(bank_messy[bank_messy['Family'].isna()])
print(bank_messy[bank_messy['CCAvg'].isna()])
print(bank_messy[bank_messy['CreditCard'].isna()])
print(bank_messy[bank_messy['Loan'].isna()])

# Handling Missing Values in the DataFrame bank_messy

# 2. Omit any row that has a variable with an NA value
bank_clean = bank_messy.dropna()

# Summary of the DataFrame bank_clean
print(bank_clean.describe(include='all'))

# 3. Impute the NA value with a value that is representative of the variable

# Impute the NA value with the previous value
bank_clean = bank_messy.fillna(method='ffill')

# Summary of the DataFrame bank_clean
print(bank_clean.describe(include='all'))
```

```python
###
### Read CSV File into a Data Frame and Show and Cleaning Outliers - Part 2
###


# Imputation using mean
bank_clean = bank_messy.copy()

bank_clean['Income'] = bank_clean['Income'].fillna(bank_clean['Income'].mean())
bank_clean['Family'] = bank_clean['Family'].fillna(bank_clean['Family'].mean())
bank_clean['CCAvg'] = bank_clean['CCAvg'].fillna(bank_clean['CCAvg'].mean())
bank_clean['CreditCard'] = bank_clean['CreditCard'].fillna(bank_clean['CreditCard'].mean())
bank_clean['Loan'] = bank_clean['Loan'].fillna(bank_clean['Loan'].mean())

# Summary of the DataFrame bank_clean
print(bank_clean.describe(include='all'))

# Imputation using median
bank_clean = bank_messy.copy()

bank_clean['Income'] = bank_clean['Income'].fillna(bank_clean['Income'].median())
bank_clean['Family'] = bank_clean['Family'].fillna(bank_clean['Family'].median())
bank_clean['CCAvg'] = bank_clean['CCAvg'].fillna(bank_clean['CCAvg'].median())
bank_clean['CreditCard'] = bank_clean['CreditCard'].fillna(bank_clean['CreditCard'].median())
bank_clean['Loan'] = bank_clean['Loan'].fillna(bank_clean['Loan'].median())


# Summary of the DataFrame bank_clean Including Objects
print(bank_clean.describe())

# Summary of the DataFrame bank_clean Including Objects
print(bank_clean.describe(include='all'))
```

```python
#### Clean the Mortage Variable

bank_messy = pd.read_csv("bank_messy.csv")  # Read the CSV file into a DataFrame

bank_clean = bank_messy.copy()  # Create a copy of the DataFrame

print(bank_clean.describe(include='all'))

# show histogram of the Mortgage variable

import matplotlib.pyplot as plt

plt.hist(bank_clean['Mortgage'], bins=20)
plt.show()

# show the boxplot of the Mortgage variable

plt.boxplot(bank_clean['Mortgage'])
plt.show

# remove outliers from the Mortgage variable

# Show the boxplot of the Mortgage variable
plt.boxplot(bank_clean['Mortgage'])
plt.show()

# Remove outliers from the Mortgage variable

Q1 = bank_clean['Mortgage'].quantile(0.25)
Q3 = bank_clean['Mortgage'].quantile(0.75)

IQR = Q3 - Q1

bank_clean = bank_clean[(bank_clean['Mortgage'] >= 100) & (bank_clean['Mortgage'] <= Q3 + 1.5 * IQR)]

print(bank_clean['Mortgage'].describe())
```

```python
#### Clean the Mortage Variable display

import matplotlib.pyplot as plt

# Use the describe command on the Mortgage column and print the result
print(bank_clean['Mortgage'].describe())

# show histogram of the Mortgage variable

plt.hist(bank_clean['Mortgage'], bins=20)
plt.show()

# show the boxplot of the Mortgage variable

plt.boxplot(bank_clean['Mortgage'])
plt.show
```

```python
#####  Create Bins for the Mortgage Variable

import pandas as pd
import matplotlib

# Create 5 bins for the variable bank_clean['Mortgage']

bank_clean['Mortgage_Bins'] = pd.cut(bank_clean['Mortgage'], bins=5, labels=False)

# create labels for the bins

bank_clean['Mortgage_Bins'] = pd.cut(bank_clean['Mortgage'], bins=5, labels=['Low', 'Low-Medium', 'Medium', 'Medium-High', 'High'])

# Show the frequency of the bins

print(bank_clean['Mortgage_Bins'].value_counts())

# Create a bar graph of the bins

plt.bar(bank_clean['Mortgage_Bins'].value_counts().index, bank_clean['Mortgage_Bins'].value_counts())
plt.show()
```

```python
  ###
  ### Basic Vector Operations - Part 1-5
  ###

import numpy as np # Importing numpy library to use numpy arrays in Python 
import pandas as pd # Importing pandas library to use pandas dataframes in Python  

# Generate a sequence of numbers from 10 to 100 with a step of 5
x = np.arange(10, 101, 5)
print(x)

# Access element from a vector:
print(x[2])  # Note: Indexing in Python starts from 0

# Display the first 4 elements
print(x[:4])

# Display the first and the third element
print(x[[0, 2]])
print(x[[0, 2, 3]])
print(x)

# Negative index (excluding elements)
print(np.delete(x, 0))  # Exclude the first element
y = np.delete(x, [0, 2])  # Exclude the first and the third element
print(y)

# Length of a vector
print(len(x))

# Sum of all elements in a vector
print(np.sum(x))

# Mean of all elements in a vector
print(np.mean(x))

# Median of all elements in a vector
print(np.median(x))

# Maximum and minimum values in a vector
print(np.max(x))
print(np.min(x))

# Sorting a vector
print(np.sort(x))

# Sorting a vector in descending order
print(-np.sort(-x))

# Handling NA and NaN values
x = np.array([1, 2, 3, 4, 5, 6, 7, 8, np.nan])
mean_x = np.nanmean(x)  # Mean excluding NaN values
print(mean_x)

# Example with NULL values
# Python does not have NULL; it has None. np.nan is used for missing values in arrays.
x = np.array([1, 2, 3, 4, 5, 6, 7, 8, None], dtype=object)
print(pd.Series(x).mean(skipna=True))  # Mean excluding None values

# Standard deviation
my_numbers = np.array([1, 2, 3, 1, 3, 5, 25])
print(np.std(my_numbers, ddof=1))  # ddof=1 for sample standard deviation

# Vector operations
print(my_numbers)
print(my_numbers * 5)
print(my_numbers + 1)
print(my_numbers + my_numbers)
```
