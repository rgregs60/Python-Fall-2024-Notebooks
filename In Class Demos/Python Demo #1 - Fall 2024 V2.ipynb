<h3>NumPy (Numerical Python)</h3>

Purpose: NumPy is a fundamental library for numerical computing in Python. 

It provides support for large, multi-dimensional arrays and matrices, along with a vast collection of mathematical functions to operate on these arrays.

* Key Features:
	o 	ndarray: The core data structure in NumPy is the ndarray, which allows for efficient storage and manipulation of numerical data.

	o 	Vectorization: NumPy enables operations on entire arrays without the need for explicit loops, which makes your code more concise and faster.

	o 	Broadcasting: This feature allows NumPy to perform operations on arrays of different shapes.
	
	o	Linear Algebra, Fourier Transform, and Random Number Capabilities: NumPy provides functions for performing linear algebra operations, 
		computing Fourier transforms, and generating random numbers.
</div>

<h3>Matplotlib.Pyplot</h3>

Purpose: is a module within the Matplotlib library in Python, which is used for creating static, interactive, and animated visualizations. It’s widely used for generating plots, charts, and figures in Python, and is particularly popular in the scientific and data science communities.

* Key Features:
	o 	Simple Interface: pyplot provides a MATLAB-like interface for creating plots, making it easy to generate plots quickly.

	o 	Versatile Plot Types: You can create various types of plots, such as line plots, bar charts, histograms, scatter plots, and more.

	o 	Customization: Almost every aspect of a plot can be customized, including titles, labels, colors, and more.
	
	o	Integration: matplotlib.pyplot integrates well with other libraries like NumPy, Pandas, and SciPy, making it an essential tool for data visualization in Python.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating an array
array = np.array([1, 2, 3, 4])

# create array2 sequential numbers from 1 to 20

# Performing an operation on the array
array_squared = array ** 2

print(array_squared)  # Output: [ 1  4  9 16]

# create a line graph of array_squared

plt.plot(array, array_squared)

plt.xlabel('Array')
plt.ylabel('Array Squared')
plt.title('Array vs Array Squared')
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter  # Import PercentFormatter

# Creating an array
array2 = np.arange(100, 2100, 100)

# Performing an operation on the array
array_squared = np.sqrt(.25 / array2) * 1.96

print(array_squared)

# Create a line graph of array_squared
plt.figure(figsize=(10, 6))  # Increase the plot size

plt.plot(array2, array_squared, color='green')  # Green plot line
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))  # Use PercentFormatter for y-axis

plt.xlabel('n')
plt.ylabel('Array Squared')
plt.title('Array vs Array Squared')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# Creating an array
array2 = np.arange(100, 2100, 100)

# Performing an operation on the array
array_squared = np.sqrt(.25 / array2) * 1.96 # 1.96 is the z-score for 95% confidence level

# Create a line graph of array_squared
plt.figure(figsize=(10, 6))  # Increase the plot size

plt.plot(array2, array_squared, color='green')  # Green plot line

# Set x-axis limits and ticks
plt.xlim(0, 2000)  # X-axis from 0 to 2000
plt.xticks(np.arange(0, 2100, 100))  # Ticks every 100

# Add vertical reference lines at 400, 900, and 1650
for x in [400, 900, 1650]:
    plt.axvline(x=x, color='gray', linestyle='--', linewidth=1)

plt.xlabel('n')
plt.ylabel('Degree of Precision')
plt.title('Sample Size and Degree of Precision at 95% Confidence Level')

plt.gca().yaxis.set_major_formatter(PercentFormatter(1))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# Creating an array
array2 = np.arange(100, 2100, 100)

# Performing an operation on the array
array_squared = np.sqrt(.25 / array2) * 1.96 # 1.96 is the z-score for 95% confidence level
array_squared2 = np.sqrt(.25 / array2) * 1.65 # 1.65 is the z-score for 90% confidence level

# Create a line graph of array_squared
plt.figure(figsize=(10, 6))  # Increase the plot size

# Create an axis object to customize the plot area
ax = plt.gca()


plt.plot(array2, array_squared, color='green')  # Green plot line
plt.plot(array2, array_squared2, color='blue')  # Blue plot line

# Set x-axis limits and ticks
plt.xlim(0, 2000)  # X-axis from 0 to 2000
plt.xticks(np.arange(0, 2100, 100))  # Ticks every 100

# Add vertical reference lines at 400, 900, and 1650
for x in [400, 900, 1650]:
    plt.axvline(x=x, color='gray', linestyle='--', linewidth=1)

plt.xlabel('n')
plt.ylabel('Degree of Precision')
plt.title('Sample Size and Degree of Precision at 90% and 95% Confidence Level')

plt.legend(['95% Confidence Level', '90% Confidence Level'])

plt.gca().yaxis.set_major_formatter(PercentFormatter(1))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

# Plotting the bar graph
plt.figure(figsize=(8, 6))
plt.bar(df['City'], df['Age'], color='skyblue')

for i, v in enumerate(df['Age']):
    plt.text(i, v + 0.5, str(v), ha='center')


plt.xlabel('City')
plt.ylabel('Age')
plt.title('Age by City')

plt.show()
```

<h3>Bar Graph</h3>
<b>"Cleaner" Bar Graph</b>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

# Plotting the bar graph
plt.figure(figsize=(8, 6))
plt.bar(df['City'], df['Age'], color='skyblue')

# Adding text labels on top of each bar
for i, v in enumerate(df['Age']):
    plt.text(i, v + 0.5, str(v), ha='center')

# Remove y-axis and borders, but keep the x-labels
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Display only the x-labels under the bars
plt.xticks(ticks=range(len(df['City'])), labels=df['City'])
plt.title('Age by City')

# Displaying the plot
plt.show()
```

```python
  ###
  ### Basic Vector Operations - Part 1-5
  ###

import numpy as np # Importing numpy library to use numpy arrays in Python 
import pandas as pd # Importing pandas library to use pandas dataframes in Python  

# Generate a sequence of numbers from 10 to 100 with a step of 5
x = np.arange(10, 101, 5)
print(x)

# Access element from a vector:
print(x[2])  # Note: Indexing in Python starts from 0

# Display the first 4 elements
print(x[:4])

# Display the first and the third element
print(x[[0, 2]])
print(x[[0, 2, 3]])
print(x)

# Negative index (excluding elements)
print(np.delete(x, 0))  # Exclude the first element
y = np.delete(x, [0, 2])  # Exclude the first and the third element
print(y)

# Length of a vector
print(len(x))

# Sum of all elements in a vector
print(np.sum(x))

# Mean of all elements in a vector
print(np.mean(x))

# Median of all elements in a vector
print(np.median(x))

# Maximum and minimum values in a vector
print(np.max(x))
print(np.min(x))

# Sorting a vector
print(np.sort(x))

# Sorting a vector in descending order
print(-np.sort(-x))

# Handling NA and NaN values
x = np.array([1, 2, 3, 4, 5, 6, 7, 8, np.nan])
mean_x = np.nanmean(x)  # Mean excluding NaN values
print(mean_x)

# Example with NULL values
# Python does not have NULL; it has None. np.nan is used for missing values in arrays.
x = np.array([1, 2, 3, 4, 5, 6, 7, 8, None], dtype=object)
print(pd.Series(x).mean(skipna=True))  # Mean excluding None values

# Standard deviation
my_numbers = np.array([1, 2, 3, 1, 3, 5, 25])
print(np.std(my_numbers, ddof=1))  # ddof=1 for sample standard deviation

# Vector operations
print(my_numbers)
print(my_numbers * 5)
print(my_numbers + 1)
print(my_numbers + my_numbers)
```
