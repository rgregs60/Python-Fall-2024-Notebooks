````python
# Check the versions of the libraries and load
import pandas as pd  #importing pandas library
import numpy as np  #importing numpy library


import matplotlib.pyplot as plt  # importing matplotlib library

# Convert the categorical columns back to numerical for calculation
reverse_labels = {v: k for k, v in labels.items()}
Sat_Data[['Performance', 'Ease_of_Use', 'Maintenance']] = Sat_Data[['Performance', 'Ease_of_Use', 'Maintenance']].replace(reverse_labels)


# remove Wave from the Bar Graph of Means

mean_values = Sat_Data.drop('Testing_Wave', axis=1).mean()

# Create a bar graph of the mean values - Remove Wave
mean_values.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Variables')
plt.ylabel('Mean Values')
plt.title('Mean Values of Performance, Ease of Use, and Maintenance')
plt.show()

# Create a cluster bar graph using waves and show the means

# Calculate the mean values for each wave
mean_values_by_wave = Sat_Data.groupby('Testing_Wave').mean()

# Plot the cluster bar graph
mean_values_by_wave.plot(kind='bar', figsize=(10, 6), color=['blue', 'green', 'red'])
plt.xlabel('Testing Wave')
plt.ylabel('Mean Values')
plt.title('Mean Values of Performance, Ease of Use, and Maintenance by Testing Wave')
plt.legend(['Performance', 'Ease of Use', 'Maintenance'])
plt.show()
````
