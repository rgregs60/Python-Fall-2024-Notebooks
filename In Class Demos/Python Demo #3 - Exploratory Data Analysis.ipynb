<h4> Output formats to Power Point </h4>
<h5> <b>PowerPoint Layouts</b> </h5>

<ol>    
    <li> 
        (0) <b>Title Slide:</b> This layout is designed for the opening slide of the presentation. It usually includes placeholders for a title and a subtitle.<br>
         <p> <p> 
        (1) <b>Title and Content:</b> A versatile layout with a title at the top and a large content area below. The content area can be used for text, tables, charts, images, and other types of content.<br>
        <p> <p>       
        (2) <b>Section Header:</b> Intended for introducing new sections within your presentation. It typically includes a title and a subtitle or text placeholder.<br>
        <p> <p> 
        (3) <b>Two Content:</b> This layout splits the slide into two main content areas, allowing for a side-by-side comparison or presentation of related information. It also includes a title placeholder.<br>

````python
# Load in additional libraries

import numpy as np
import pandas as pd
import statsmodels.api as sm  # Import the statsmodels.api module
from statsmodels.formula.api import ols  # Import the ols function from statsmodels.formula.api module
from statsmodels.stats.multicomp import pairwise_tukeyhsd  # Import the pairwise_tukeyhsd function
from sklearn.preprocessing import LabelEncoder  # Optional, but not necessary for ANOVA

# Set random seed for reproducibility
np.random.seed(0)

# Independent variables
IV1 = ['A', 'B']
IV2 = ['X', 'Y']

# Generate random data
data = {
    'IV1': np.random.choice(IV1, 100),  # Categorical data for IV1
    'IV2': np.random.choice(IV2, 100),  # Categorical data for IV2
    'DV': np.random.normal(0, 1, 100)   # Normally distributed dependent variable
}

# Create DataFrame
df_anova = pd.DataFrame(data)

# Display data types
print(df_anova.dtypes)

# Show a summary table of responses for each unique value of IV1 and IV2
print(df_anova.groupby(['IV1', 'IV2'])['DV'].describe())

# Perform two-way ANOVA using OLS
model = ols('DV ~ C(IV1) * C(IV2)', data=df_anova).fit()  # C(IV1) and C(IV2) indicate categorical variables
anova_table = sm.stats.anova_lm(model, typ=2)  # Perform ANOVA (Type II)
print(anova_table)

# Perform Tukey's HSD test if needed
tukey_results = pairwise_tukeyhsd(df_anova['DV'], df_anova['IV1'] + df_anova['IV2'])
print(tukey_results)
````

<h4>ANOVA is a statistical technique used to determine if there are any statistically significant differences between the means of three or more independent (unrelated) groups.</h4>

<p>Key points about ANOVA:</p>
<p> <p>
	<ol>
		<il>
		•	Purpose: To test for significant differences among group means.<br>
		•	Types:<br>
		•	One-Way ANOVA: Compares means among three or more groups based on one factor.<br>
		•	Two-Way ANOVA: Compares means based on two factors and can include interaction effects between factors.<br>
		•	Assumptions:<br>
		•	Independence of observations.<br>
		•	Normally distributed populations.<br>
		•	Homogeneity of variances (equal variances among groups).<br>
		•	Output: The main output is the F-statistic and its associated p-value. <br>
			<p>  <p>
			If the p-value is less than the significance level (e.g., 0.05), we reject <br>
			the null hypothesis that all group means are equal.<br>
		</il>
	</ol>
	<p> <p>
	<ol>
	<p>Tukey’s HSD (Honestly Significant Difference) Test<p>
	<p> <p>
	<p> Tukey’s HSD is a post-hoc test used after an ANOVA to find out which specific group means are different. It controls for Type I error across multiple comparisons.<p>
	<p> <p>
	<p>Key points about Tukey’s HSD:<p>
	<p> <p>
		<il>
		•	Purpose: To identify which pairs of group means are significantly different.
		•	Usage: Performed after a significant ANOVA result.
		•	Assumptions: Same as ANOVA (independence, normality, homogeneity of variances).
		•	Output: Provides confidence intervals and p-values for the differences between each pair of group means.
		<il>
	<p>Example Workflow<p>
	<p> <p>
		<il>
		1.	Conduct ANOVA: Check if there are any significant differences among the group means.
		2.	Post-Hoc Analysis (if ANOVA is significant): Perform Tukey’s HSD to identify which specific groups differ.</p>
		</il>
	</ol>
</div>

<h3> Creating and Using Functions for Repeated Tasks<h3>
</div>
<p>Instead of Repeating the Coding, we can define a Function</p>
<p>There are a number of arguments passed by the function which allows to have more control over the output</p>



````python
##### Create Regression Function

def plot_regression_with_r_squared(df, x_var, y_var, figsize=(10, 6), style="darkgrid", ylim=(0, 50000), title=""):
    import matplotlib.pyplot as plt
    import seaborn as sns
    from scipy import stats
    
    plt.figure(figsize=figsize)
    sns.set_style(style)
    sns.regplot(x=x_var, y=y_var, data=df)
    plt.ylim(ylim)  # Optionally, specify an upper limit as plt.ylim(0, upper_limit)
    plt.title(title)

    # Calculate R-squared value
    slope, intercept, r_value, p_value, std_err = stats.linregress(df[x_var], df[y_var])
    r_squared = r_value**2

    # Display R-squared value on the plot
    plt.text(0.05, 0.95, f'R² = {r_squared:.2f}', transform=plt.gca().transAxes)

    plt.show()
````

````python
### Call the Function Created

plot_regression_with_r_squared(df, 'horsepower','price', figsize=(10, 6), style="darkgrid", ylim=(0, 50000), title="Price vs. Horsepower")
plot_regression_with_r_squared(df, 'stroke','price', figsize=(10, 6), style="darkgrid", ylim=(0, 50000), title="Stroke vs. Price")
plot_regression_with_r_squared(df, 'stroke','horsepower', figsize=(10, 6), style="darkgrid", ylim=(0), title="Stroke vs. Horsepower")
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os

"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Performed 4 aggregations grouped on columns: 'make', 'body-style'
    df = df.groupby(['make', 'body-style']).agg(horsepower_mean=('horsepower', 'mean'), price_mean=('price', 'mean'), citympg_mean=('city-mpg', 'mean'), highwaympg_mean=('highway-mpg', 'mean')).reset_index()
    return df

df_clean = clean_data(df.copy())
df_clean.head()

# export df_clean to excel File   

print(os.getcwd()) # Output: /
df_clean.to_excel('df_clean.xlsx')

````

<h2>Section 4: Correlation and Causation</h2>
</div>

<p><b>Correlation</b>: a measure of the extent of interdependence between variables.</p>

<p><b>Causation</b>: the relationship between cause and effect between two variables.</p>

<p>It is important to know the difference between these two. Correlation does not imply causation. Determining correlation is much simpler  the determining causation as causation may require independent experimentation.</p>


<p><b>Pearson Correlation</b></p>
<p>The Pearson Correlation measures the linear dependence between two variables X and Y.</p>
<p>The resulting coefficient is a value between -1 and 1 inclusive, where:</p>
<ul>
    <li><b>1</b>: Perfect positive linear correlation.</li>
    <li><b>0</b>: No linear correlation, the two variables most likely do not affect each other.</li>
    <li><b>-1</b>: Perfect negative linear correlation.</li>
</ul>


<p>Pearson Correlation is the default method of the function "corr". Like before, we can calculate the Pearson Correlation of the of the 'int64' or 'float64'  variables.</p>


````python
import pandas as pd
import numpy as np

# Example DataFrame creation (remove this if df is already defined)
df = pd.DataFrame({
    'stroke': np.random.rand(100),
    'price': np.random.rand(100) * 50000
})

# Calculate the correlation matrix
correlation_matrix = df.corr()

print(correlation_matrix)
````

Sometimes we would like to know the significant of the correlation estimate. 


<b>P-value</b>
<p>What is this P-value? The P-value is the probability value that the correlation between these two variables is statistically significant. Normally, we choose a significance level of 0.05, which means that we are 95% confident that the correlation between the variables is significant.</p>

By convention, when the
<ul>
    <li>p-value is < 0.001: we say there is strong evidence that the correlation is significant.</li>
    <li>the p-value is < 0.05: there is moderate evidence that the correlation is significant.</li>
    <li>the p-value is < 0.1: there is weak evidence that the correlation is significant.</li>
    <li>the p-value is > 0.1: there is little evidence that the correlation is significant.</li>
</ul>


 We can obtain this information using  "stats" module in the "scipy"  library.


````python
from scipy import stats
````

<h4>ANOVA: Analysis of Variance</h4>
<p>The Analysis of Variance  (ANOVA) is a statistical method used to test whether there are significant differences between the means of two or more groups. ANOVA returns two parameters:</p>

<p><b>F-test score</b>: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.</p>

<p><b>P-value</b>:  P-value tells how statistically significant our calculated score value is.</p>

<p>If our price variable is strongly correlated with the variable we are analyzing, we expect ANOVA to return a sizeable F-test score and a small p-value.</p>
</div>
